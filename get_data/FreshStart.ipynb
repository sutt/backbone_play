```python
### Given Functions ####
def skew_increments(s):
    return [{'C': -1, 'A':0, 'T':0, 'G':1, 
             'N':0, 'S':0,'R':0, 'D':0, 'Y':0, 
             'K':0, 'M':0, 'W':0, 'B':0, 'V':0, 
             'H':0}[c] for c in s]
def gc_skew(s):
    return numpy.cumsum([0]+skew_increments(s))
```

```python
%time gc1 = gc_skew(seq)
```

```python
def loop_moving_avg(s, smoothing):
    
    y = []
    print type(s)
    half_smooth = int(round(smoothing / 2,0))
    
    for i in range(len(s)):
        
        if ( (i > half_smooth) and (i < (len(s)- half_smooth)) ):
            y.append( int(
                     sum(s[range( i - half_smooth , i +half_smooth)]) / smoothing
                     ) )
        else:
            y.append(s[i])
        
    return y
```

```python
%time test = loop_moving_avg(gc1[0:100000],100)
print type(test)
```

```python
def vector_moving_avg(s, smoothing):
    
    y = []
    print type(s)
    half_smooth = int(round(smoothing / 2,0))
    
    return [  
            sum(s[range( i - half_smooth , i +half_smooth)]) / smoothing
            for i in numpy.arange(half_smooth,len(s) - half_smooth,1)
            ]
```

```python
%time test = vector_moving_avg(gc1[0:100000],100)
print type(test)
```

```python
def numpy_moving_avg(s, smoothing):
    
    y = []
    print type(s)
    half_smooth = int(round(smoothing / 2,0))
    
    return [  
            numpy.sum(s[range( i - half_smooth , i + half_smooth)]) / smoothing
#             numpy.average(s[range( i - half_smooth , i +half_smooth)])
            for i in numpy.arange(half_smooth,len(s) - half_smooth,1)
            ]
```

```python
%time test = numpy_moving_avg(gc1[0:100000],100)
print type(test)
```

```python
### Test scaling for smooth paramerter ###

# Results: time[numpy] = 1/n * time[vector] | smooth = 10^n, n in [1,3]

gc2 = gc1[0:100000]
for i in range(1,4,1):
    print 'smooth = ', (10**i)
    %time test = vector_moving_avg(gc2, (10**i) )
    %time test = numpy_moving_avg(gc2, (10**i) )
```

```python
def sample_moving_avg(s,smoothing,sample):
    
    half_smooth = int(round(smoothing / 2,0))
    
#     y = numpy.array(len(s))
    
    return [
            numpy.sum( s[range( i - half_smooth, i + half_smooth) ] )
            for i in numpy.arange(sample + max(half_smooth,sample) , len(s)- max(half_smooth,sample), sample)
           ]
```
